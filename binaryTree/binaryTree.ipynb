```python
class TreeNode():
    def __init__(self):
        self.left = None
        self.data = None
        self.right = None
    
node1 = TreeNode()
node1.data = '호날두'

node2 = TreeNode()
node2.data = '메시'
node1.left = node2

node3 = TreeNode()
node3.data = '박관호'
node1.right = node3

node4 = TreeNode()
node4.data = '노무현'
node2.left = node4

node5 = TreeNode()
node5.data = '이명박'
node2.right = node5

node6 = TreeNode()
node6.data = '문재인'
node3.left = node6

def preorder(node):
    if node == None:
        return
    print(node.data, end=' -> ')
    preorder(node.left)
    preorder(node.right)
    preorder(node.left)

def inorder(node):
    if node == None:
        return
    inorder(node.left)
    print(node.data, end=' -> ')
    inorder(node.right)

def postorder(node):
    if node == None:
        return
    postorder(node.left)
    postorder(node.right)
    print(node.data, end=' -> ')

print('전위 순회 : ', end=' ')
preorder(node1)
print('끝')

print('중위 순회 : ', end=' ')
inorder(node1)
print('끝')

print('후위 순회 : ', end=' ')
postorder(node1)
print('끝')
```

```python
class TreeNode():
    def __init__(self):
        self.left = None
        self.data = None
        self.right = None

memory = []
root = None
nameArr = ['호날두', '메시', '박관호', '노무현', '이명박', '문재인']

node = TreeNode()
node.data = nameArr[0]
root = node
memory.append(node)

for name in nameArr[1:]:
    node = TreeNode()
    node.data = name

    current = root
    while True:
        if name < current.data:
            if current.left == None:
                current.left = node
                break
            current = current.left
        else:
            if current.right == None:
                current.right = node
                break
            current = current.right

    memory.append(node)

nameToFind = input('찾을 이름 입력')
current = root
while True:
    if nameToFind == current.data:
        print(f'{nameToFind} 찾음')
        break
    elif nameToFind < current.data:
        if current.left == None:
            print(f'트리에 {nameToFind} 없음')
            break
        current = current.left
    else:
        if current.right == None:
            print(f'트리에 {nameToFind} 없음')
            break
        current = current.right

nameToDel = input('삭제할 이름 입력')

current = root
parent = None
while True:
    if nameToDel == current.data:
        if current.left == None and current.right == None:
            if parent == None:
                root = None
                del(root)
                print('최상위 트리 삭제됨')
                break
            elif parent.left == current:
                parent.left = None
            else:
                parent.right = None
            del(current)
        elif current.left != None and current.right == None:
            if parent == None:
                root = None
                del(root)
                print('최상위 트리 삭제됨')
                break
            elif parent.left == current:
                parent.left = current.left
            else:
                parent.right = current.left
            del(current)
        elif current.left == None and current.right != None:
            if parent == None:
                root = None
                del(root)
                print('최상위 트리 삭제됨')
                break
            elif parent.left == current:
                parent.left = current.right
            else:
                parent.right = current.right
            del(current)

        print(f'{nameToDel} 삭제됨')
        break
    elif nameToDel < current.data:
        if current.left == None:
            print(f'트리에 {nameToDel} 없음')
            break
        parent = current
        current = current.left
    else:
        if current.right == None:
            print(f'트리에 {nameToDel} 없음')
            break
        parent = current
        current = current.right
```
